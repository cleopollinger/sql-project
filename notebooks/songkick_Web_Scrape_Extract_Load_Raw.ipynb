```python
import requests
from bs4 import BeautifulSoup

url = "https://www.songkick.com/metro-areas/17835-us-los-angeles-la"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

concerts = soup.select('li.event-listings-element')

concert_details = {
    'artist': [],
    'date': [],
    'venue': []
}

for concert in concerts:
    try:
        artist_tag = concert.select_one('div.artists-venue-location-wrapper strong')
        date_tag = concert.select_one('time')
        venue_tag = concert.select_one('p.location')

        artist = artist_tag.text.strip() if artist_tag else "N/A"
        date = date_tag['datetime'] if date_tag else "N/A"
        venue = venue_tag.text.strip() if venue_tag else "N/A"

        print(f"Artist: {artist}")
        print(f"Date: {date}")
        print(f"Venue: {venue}")
        print("------------------------------------------------")

        concert_details['artist'].append(artist)
        concert_details['date'].append(date)
        concert_details['venue'].append(venue)

    except Exception as e:
        print(f"Error parsing concert: {e}")

```
